```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/Cleaned_Food_Dataset 1.csv', on_bad_lines='skip')
df = pd.DataFrame(data)
df.head(10)
```

```python
def find_recipe(food_name):
    """
    Finds and displays the details of a recipe based on its name.
    """
    # Search for the food name in the 'TranslatedRecipeName' column.
    # We use .str.contains() for a flexible, case-insensitive search.
    recipe = df[df['TranslatedRecipeName'].str.contains(food_name, case=False, na=False)]

    # Check if any recipe was found
    if not recipe.empty:
        # Get the details of the first match found
        name = recipe.iloc[0]['TranslatedRecipeName']
        ingredients = recipe.iloc[0]['TranslatedIngredients']
        instructions = recipe.iloc[0]['TranslatedInstructions']
        # url = recipe.iloc[0]['URL'] # Removed as 'URL' column does not exist

        # Print the details in a clean format
        print(f"--- Recipe Found: {name} ---")
        print("\nIngredients:")
        print(ingredients)
        print("\nInstructions:")
        print(instructions)
        # print(f"\nURL: {url}") # Removed as 'URL' column does not exist

    else:
        print(f"Sorry, I couldn't find a recipe named '{food_name}'. Please check the spelling or try another name.")
```

```python
# --- Get User Input for Food Name ---

# Ask the user to enter the name of the food
user_food_name = input("Enter the exact name of the recipe you are looking for: ")

# Call the function to find and display the ingredients
find_recipe(user_food_name)
```

```python
import pickle

# Define the filename to save the pickled object
food_recipe = 'dataframe.pkl'

# Open the file in binary write mode and save the object
with open(food_recipe, 'wb') as f:
    pickle.dump(df, f)

print(f"DataFrame saved successfully to {food_recipe}")
```
